### Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import/Read Data Set

```python
url ='https://raw.githubusercontent.com/kxjw98/factor/main/sv%20-%20Sheet1.csv'
df = pd.read_csv(url)
df.head()
df.describe()
```

```python
df.describe()
```

### Multiple Linear Regression

```python
#define x and y

#total_accidents
x=df.drop(['total_acci'], axis=1).values
y=df['total_acci'].values
```

```python
print(x)
print(y)
```

```python
#Split the data set in training set and test set 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.35,random_state=0)
```

```python
#train the model on the training set 
from sklearn.linear_model import LinearRegression 
ml=LinearRegression()
ml.fit(x_train,y_train)
```

```python
#Predctions
y_pred=ml.predict(x_test)
print(y_pred)

plt.scatter(y_test, y_pred, color='red')
plt.title('Prediciones')

pred_y_df=pd.DataFrame({'Predicted Values':y_pred})
pred_y_df
pred_y_df.describe()
```
